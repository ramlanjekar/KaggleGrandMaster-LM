```python
#@title Kaggle Slug to code

import kaggle
from kaggle.api.kaggle_api_extended import KaggleApi
import json
import os
import zipfile # Import zipfile
from typing import Dict, List, Optional

class KaggleScoreSortedExtractor:
    def __init__(self):
        """Initialize Kaggle API client"""
        self.api = KaggleApi()
        self.api.authenticate()
        self.counter = 0;
        self.notebook_list =[]

        print("API Authentication sucessful")

    def get_best_unpinned_notebook(self, competition_name: str) -> Optional[Dict]:
        """
        Get the first unpinned notebook sorted by public score (highest first)

        Args:
            competition_name: Name of the competition

        Returns:
            Best unpinned notebook data with code
        """

        try:
              print(f"Fetching notebooks for: {competition_name}")
              print("Sorting by public score (descending)...")

              # Get notebooks sorted by score (highest first)
              notebooks = self.api.kernels_list(
                  competition=competition_name,
                  sort_by="scoreDescending",  # This should sort by public score
                  page_size=10 # Get top 50 to have options
              )

              print(f"Found {len(notebooks)} notebooks")

              if not notebooks:
                  return {"error": "No notebooks found"}

              # Look for first unpinned notebook
              for i, notebook in enumerate(notebooks):
                  is_pinned = getattr(notebook, 'isPinned', False)
                  high_engagement = getattr(notebook, 'voteCount', 0) > 100 and i < 3
                  title_indicators = any(indicator in notebook.title.lower() for indicator in
                                        ['getting started', 'starter', 'baseline', 'introduction'])
                  likely_pinned = is_pinned or (high_engagement and title_indicators)

                  if likely_pinned:
                      print(f"  ⚠️  Likely pinned - skipping")
                      continue
                  else:
                      print(f"  ✅ First unpinned notebook found!")

                      # Download and extract code
                      notebook_data = self.download_and_extract_code(notebook)
                      if "error" not in notebook_data: # Check if download/extraction was successful
                          return notebook_data

                      else:
                          print(f"Skipping this notebook due to download/extraction error: {notebook_data['error']}")
                          continue # Try the next notebook if this one failed to download/extract

              return {"error": "No unpinned notebooks found or all failed to download/extract"}

        except Exception as e:
              print(f"Error: {e}")
              return {"error": str(e)}

        return self.notebook_list

    def download_and_extract_code(self, notebook) -> Dict:
        """
        Download notebook and extract code

        Args:
            notebook: Kaggle notebook object

        Returns:
            Dictionary with notebook data and code
        """
        temp_dir = "./temp_notebook/"
        notebook_path = None # Initialize notebook_path
        try:
            # Create temp directory
            os.makedirs(temp_dir, exist_ok=True)

            # Define the expected path for the downloaded ZIP file
            # Kaggle API's kernels_pull downloads a ZIP named after the kernel reference
            zip_file_name = f"{notebook.ref.replace('/', '_')}.zip"
            zip_file_path = os.path.join(temp_dir, zip_file_name)

            print(f"Downloading: {notebook.ref}")
            # The kernels_pull command downloads the ZIP to the specified path
            self.api.kernels_pull(notebook.ref, path=temp_dir, metadata=True)

            # After download, check if the zip file exists
            if not os.path.exists(zip_file_path):
                # If the zip wasn't downloaded as expected, it might have extracted directly
                # (less common, but possible for older API versions or specific kernel types)
                # In this case, we proceed to check for .ipynb directly, but the primary expectation is a zip.
                print(f"DEBUG: ZIP file not found at {zip_file_path}. Checking for direct .ipynb extraction.")
            else:
                print(f"DEBUG: ZIP file downloaded to: {zip_file_path}")
                # Unzip the downloaded file
                with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                    zip_ref.extractall(temp_dir)
                print(f"DEBUG: Extracted contents of {zip_file_path} to {temp_dir}")
                os.remove(zip_file_path) # Clean up the zip file

            # Find notebook file (now that it's extracted from the zip)
            notebook_files = [f for f in os.listdir(temp_dir) if f.endswith('.ipynb')]

            if not notebook_files:
                print(f"DEBUG: No .ipynb file found in {temp_dir} after extraction.")
                return {"error": "No notebook file found after download/extraction"}

            notebook_path = os.path.join(temp_dir, notebook_files[0])
            print(f"DEBUG: Attempting to open notebook at: {notebook_path}")
            print(f"DEBUG: File exists: {os.path.exists(notebook_path)}")
            if os.path.exists(notebook_path):
                print(f"DEBUG: File size: {os.path.getsize(notebook_path)} bytes")
            else:
                print("DEBUG: File does not exist after path construction.")

            # Extract code
            with open(notebook_path, 'r', encoding='utf-8') as f:
                notebook_data = json.load(f)

            # Get all code cells
            code_cells = []
            for cell in notebook_data.get('cells', []):
                if cell.get('cell_type') == 'code':
                    source = cell.get('source', [])
                    if isinstance(source, list):
                        code_cells.append(''.join(source))
                    else:
                        code_cells.append(source)

            full_code = '\n\n# ' + '='*80 + '\n# NEW CELL\n# ' + '='*80 + '\n\n'.join(code_cells)

            # Return structured data
            result = {
                "competition": notebook.ref.split('/')[0] if '/' in notebook.ref else "unknown",
                "notebook_ref": notebook.ref,
                "title": notebook.title,
                "author": notebook.author,
                "public_score": getattr(notebook, 'publicScore', 'N/A'),
                "vote_count": getattr(notebook, 'voteCount', 'N/A'),
                "language": getattr(notebook, 'language', 'Python'),
                "code": full_code,
                "code_length": len(full_code),
                "num_cells": len(code_cells)
            }

            return result

        except json.JSONDecodeError as e:
            print(f"Error parsing JSON from notebook: {e}")
            # If you were expecting a .ipynb but got something else,
            # it might be a script or a malformed notebook.
            # You can add logic here to handle non-.ipynb files if necessary.
            return {"error": f"Invalid notebook file format: {str(e)}. The file might be a script, not a Jupyter notebook, or corrupted."}
        except Exception as e:
            print(f"Error downloading/extracting: {e}")
            return {"error": f"Download failed: {str(e)}"}
        finally:
            # Clean up the temporary directory and its contents
            if os.path.exists(temp_dir):
                for item in os.listdir(temp_dir):
                    item_path = os.path.join(temp_dir, item)
                    try:
                        if os.path.isfile(item_path):
                            os.remove(item_path)
                        elif os.path.isdir(item_path):
                            os.rmdir(item_path) # or shutil.rmtree for non-empty dirs
                    except Exception as cleanup_e:
                        print(f"Warning: Could not clean up {item_path}: {cleanup_e}")
                try:
                    os.rmdir(temp_dir)
                except OSError as e: # Catch OSError for non-empty directory
                    print(f"Warning: Could not remove directory {temp_dir}: {e}. It might not be empty.")


# Test function for single competition
def test_single_competition():
    """Test the extractor with ISIC 2024 competition"""

    extractor = KaggleScoreSortedExtractor()

    # Competition name (use the actual Kaggle slug)
    competition_name = "titanic"

    print(f"{'='*80}")
    print(f"TESTING: {competition_name}")
    print(f"{'='*80}")

    # Get best unpinned notebook
    result = extractor.get_best_unpinned_notebook(competition_name)

    # Display results
    if "error" in result:
        print(f"\n❌ Error: {result['error']}")
    else:
        print(f"\n✅ SUCCESS!")
        print(f"📊 Competition: {competition_name}")
        print(f"📝 Notebook: {result['notebook_ref']}")
        print(f"🏆 Title: {result['title']}")
        print(f"👤 Author: {result['author']}")
        print(f"📈 Public Score: {result['public_score']}")
        print(f"👍 Votes: {result['vote_count']}")
        print(f"💻 Language: {result['language']}")
        print(f"📄 Code Length: {result['code_length']} characters")
        print(f"🔢 Number of Cells: {result['num_cells']}")

        print(f"\n{'='*80}")
        print(f"EXTRACTED CODE:")
        print(f"{'='*80}")

        # Show first 2000 characters of code
        code_preview = result['code'][:2000]
        print(code_preview)
        if len(result['code']) > 2000:
            print(f"\n... [Code truncated - Total length: {result['code_length']} characters]")

if __name__ == "__main__":
    test_single_competition()
```

```python
#@title Competition name -> Slug

import re

def get_slug(competition_name):
  comps = api.competitions_list(search=competition_name)

  if not comps:
    return None

  url_text = comps[0].ref

  pattern = r"https://www\.kaggle\.com/(?:competitions|c)/([^/]+)"

  match = re.search(pattern, url_text)

  if match:
      extracted_text = match.group(1)
      return extracted_text

get_slug("Bike Sharing Demand")
```

```python
#@title Data Fetching Pipeline

import kaggle
from kaggle.api.kaggle_api_extended import KaggleApi
import json
import os
import zipfile # Import zipfile
from typing import Dict, List, Optional
import asyncio
from tqdm import tqdm
from time import time

async def DataProcessing(df_competitions):
    class KaggleScoreSortedExtractor:
        def __init__(self,output_dir: str = '/content/drive/MyDrive/KaggleGMData'):
            """Initialize Kaggle API client"""
            self.api = KaggleApi()
            self.api.authenticate()
            self.output_dir = output_dir
            self.jsonl_file = os.path.join(self.output_dir, "kaggle_notebooks.jsonl")

            print("API Authentication sucessful")

        async def get_best_unpinned_notebook(self, competition_name: str,semaphore:asyncio.Semaphore) -> Optional[Dict]:
            """
            Get the first unpinned notebook sorted by public score (highest first)

            Args:
                competition_name: Name of the competition

            Returns:
                Best unpinned notebook data with code
            """
            async with semaphore:
                try:
                      print(f"Fetching notebooks for: {competition_name}")
                      print("Sorting by public score (descending)...")

                      # Get notebooks sorted by score (highest first)
                      notebooks = self.api.kernels_list(
                          competition=competition_name,
                          sort_by="scoreDescending",  # This should sort by public score
                          page_size=10 # Get top 50 to have options
                      )

                      print(f"Found {len(notebooks)} notebooks")

                      if not notebooks:
                          return {"error": "No notebooks found"}

                      # Look for first unpinned notebook
                      for i, notebook in enumerate(notebooks):
                          is_pinned = getattr(notebook, 'isPinned', False)
                          high_engagement = getattr(notebook, 'voteCount', 0) > 100 and i < 3
                          title_indicators = any(indicator in notebook.title.lower() for indicator in
                                                ['getting started', 'starter', 'baseline', 'introduction'])
                          likely_pinned = is_pinned or (high_engagement and title_indicators)

                          if likely_pinned:
                              print(f"  ⚠️  Likely pinned - skipping")
                              continue
                          else:
                              print(f"  ✅ First unpinned notebook found!")

                              # Download and extract code
                              notebook_data = await self.download_and_extract_code(notebook)
                              if "error" not in notebook_data: # Check if download/extraction was successful
                                  notebook_data["competition"] = competition_name
                                  self.append_to_jsonl(notebook_data)
                                  return notebook_data

                              else:
                                  print(f"Skipping this notebook due to download/extraction error: {notebook_data['error']}")
                                  continue # Try the next notebook if this one failed to download/extract

                      return {"error": "No unpinned notebooks found or all failed to download/extract"}

                except Exception as e:
                  print(f"Error: {e}")
                  return {"error": str(e)}

            return self.notebook_list

        def append_to_jsonl(self,data):
            with open(self.jsonl_file,'a',encoding = 'utf-8') as f:
                json.dump(data,f,ensure_ascii = False)
                f.write('\n')


        async def download_and_extract_code(self, notebook) -> Dict:
            """
            Download notebook and extract code

            Args:
                notebook: Kaggle notebook object

            Returns:
                Dictionary with notebook data and code
            """
            temp_dir = "./temp_notebook/"
            notebook_path = None # Initialize notebook_path
            try:
                # Create temp directory
                os.makedirs(temp_dir, exist_ok=True)

                # Define the expected path for the downloaded ZIP file
                # Kaggle API's kernels_pull downloads a ZIP named after the kernel reference
                zip_file_name = f"{notebook.ref.replace('/', '_')}.zip"
                zip_file_path = os.path.join(temp_dir, zip_file_name)

                print(f"Downloading: {notebook.ref}")
                # The kernels_pull command downloads the ZIP to the specified path
                self.api.kernels_pull(notebook.ref, path=temp_dir, metadata=True)

                # After download, check if the zip file exists
                if not os.path.exists(zip_file_path):
                    # If the zip wasn't downloaded as expected, it might have extracted directly
                    # (less common, but possible for older API versions or specific kernel types)
                    # In this case, we proceed to check for .ipynb directly, but the primary expectation is a zip.
                    print(f"DEBUG: ZIP file not found at {zip_file_path}. Checking for direct .ipynb extraction.")
                else:
                    print(f"DEBUG: ZIP file downloaded to: {zip_file_path}")
                    # Unzip the downloaded file
                    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                        zip_ref.extractall(temp_dir)
                    print(f"DEBUG: Extracted contents of {zip_file_path} to {temp_dir}")
                    os.remove(zip_file_path) # Clean up the zip file

                # Find notebook file (now that it's extracted from the zip)
                notebook_files = [f for f in os.listdir(temp_dir) if f.endswith('.ipynb')]

                if not notebook_files:
                    print(f"DEBUG: No .ipynb file found in {temp_dir} after extraction.")
                    return {"error": "No notebook file found after download/extraction"}

                notebook_path = os.path.join(temp_dir, notebook_files[0])
                print(f"DEBUG: Attempting to open notebook at: {notebook_path}")
                print(f"DEBUG: File exists: {os.path.exists(notebook_path)}")
                if os.path.exists(notebook_path):
                    print(f"DEBUG: File size: {os.path.getsize(notebook_path)} bytes")
                else:
                    print("DEBUG: File does not exist after path construction.")

                # Extract code
                with open(notebook_path, 'r', encoding='utf-8') as f:
                    notebook_data = json.load(f)

                # Get all code cells
                code_cells = []
                for cell in notebook_data.get('cells', []):
                    if cell.get('cell_type') == 'code':
                        source = cell.get('source', [])
                        if isinstance(source, list):
                            code_cells.append(''.join(source))
                        else:
                            code_cells.append(source)

                full_code = '\n\n# ' + '='*80 + '\n# NEW CELL\n# ' + '='*80 + '\n\n'.join(code_cells)

                # Return structured data
                result = {
                    "competition": notebook.ref.split('/')[0] if '/' in notebook.ref else "unknown",
                    "notebook_ref": notebook.ref,
                    "title": notebook.title,
                    "author": notebook.author,
                    "public_score": getattr(notebook, 'publicScore', 'N/A'),
                    "vote_count": getattr(notebook, 'voteCount', 'N/A'),
                    "language": getattr(notebook, 'language', 'Python'),
                    "code": full_code,
                    "code_length": len(full_code),
                    "num_cells": len(code_cells)
                }

                return result

            except json.JSONDecodeError as e:
                print(f"Error parsing JSON from notebook: {e}")
                # If you were expecting a .ipynb but got something else,
                # it might be a script or a malformed notebook.
                # You can add logic here to handle non-.ipynb files if necessary.
                return {"error": f"Invalid notebook file format: {str(e)}. The file might be a script, not a Jupyter notebook, or corrupted."}
            except Exception as e:
                print(f"Error downloading/extracting: {e}")
                return {"error": f"Download failed: {str(e)}"}
            finally:
                # Clean up the temporary directory and its contents
                if os.path.exists(temp_dir):
                    for item in os.listdir(temp_dir):
                        item_path = os.path.join(temp_dir, item)
                        try:
                            if os.path.isfile(item_path):
                                os.remove(item_path)
                            elif os.path.isdir(item_path):
                                os.rmdir(item_path) # or shutil.rmtree for non-empty dirs
                        except Exception as cleanup_e:
                            print(f"Warning: Could not clean up {item_path}: {cleanup_e}")
                    try:
                        os.rmdir(temp_dir)
                    except OSError as e: # Catch OSError for non-empty directory
                        print(f"Warning: Could not remove directory {temp_dir}: {e}. It might not be empty.")

    async def process_competitions(slugs:List[str],batch_size: int = 4):
      extractor = KaggleScoreSortedExtractor()
      semaphore = asyncio.Semaphore(batch_size)

      for i in tqdm(range(0,len(slugs),batch_size)):
          start = time()
          batch = slugs[i : i+batch_size]
          tasks = [extractor.get_best_unpinned_notebook(slug,semaphore) for slug in batch]
          results = await asyncio.gather(*tasks,return_exceptions = True)

          for slug, result in zip(batch, results):
              if isinstance(result, Exception):
                  df_competitions.loc[df_competitions['Slug'] == slug, 'Status'] = 'Error'
                  df_competitions.loc[df_competitions['Slug'] == slug, 'ErrorMessage'] = str(result)
              elif "error" in result:
                  df_competitions.loc[df_competitions['Slug'] == slug, 'Status'] = 'Error'
                  df_competitions.loc[df_competitions['Slug'] == slug, 'ErrorMessage'] = result["error"]
              else:
                  df_competitions.loc[df_competitions['Slug'] == slug, 'Status'] = 'Success'

          # After each batch
          dir_path = os.path.join(os.getcwd(), 'KaggleGMData')
          os.makedirs(dir_path, exist_ok=True)

          path = os.path.join(dir_path, 'competition.csv')
          df_competitions.to_csv(path, index=False)

          await asyncio.sleep(10)


          end = time()

          print(f"Batch {i} took {end - start} seconds\n\n")


    async def main():
        await process_competitions(df_competitions['Slug'].tolist())
        print("\n Data fetching Sucessful !!!!")


    await main()
```
